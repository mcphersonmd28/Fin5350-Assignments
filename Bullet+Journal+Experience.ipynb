## Keeping a Daily Bullet Journal

I decided to keep a bullet journal during the class as suggested by Dr. Brough.  I was glad that I chose to keep a bullet journal because it made me more organized and focused.  I had heard about bullet journals before but I never really understood what they were until I researched them more.  The bullet journals I had seen before were very creative and artistic, but when I was researching them, I found that they could be very simple, almost like a day planner.

As I went throughout the semester, I saw that I was keeping track what I needed to do and my assignments better than I had before.  There were some days that I did not write in my bullet journal.  During these days, I noticed that I was not as productive.  Keeping a bullet journal requires dedication, but, in the end, the dedication pays off.  I found that keeping a bullet journal was a really good experience and I plan on keeping one for next semester as well.
